Step 3 Preprocess Data

Step 4 Feature Extraction

```python
import pandas as pd
def getdata(dataFile):
    tr = pd.read_csv(dataFile)
    fpk = tr[['First_pokemon']].join(pk.set_index('#'),on='First_pokemon')
    spk = tr[['Second_pokemon']].join(pk.set_index('#'),on='Second_pokemon')
    fpk_v = fpk._get_numeric_data().iloc[:,1:]
    spk_v = spk._get_numeric_data().iloc[:,1:]
    fpk_v['Legendary'] = fpk_v['Legendary'].astype('int')
    spk_v['Legendary'] = spk_v['Legendary'].astype('int')
    feat  = fpk_v - spk_v
    lab = (tr['First_pokemon'] == tr['Winner']).astype('int')
    return feat,lab
```

Step 5 Model train

```python
from sklearn.ensemble import GradientBoostingClassifier
def trainclassifer(feat,label):
    mod = GradientBoostingClassifier()
    mod.fit(feat,label)
    return mod
```

Step 6 Model evaluate

```python
def evaluation_accuracy(prediction,actual):
    acc = sum(prediction==actual)/len(prediction)
    return acc
```

Step 7 Run pipeline

```python
def pipeline():
    trainFeat,trainLab = getdata('train.csv')
    testFeat,testLab = getdata('test.csv')
    mod = trainclassifer(trainFeat,trainLab)
    res = mod.predict(testFeat)
    acc = evaluation_accuracy(res,testLab)
    return acc
```

```python
pipeline()
```
